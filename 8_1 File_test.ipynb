```python
# 김소월 진달래꽃 - 파일에 write, read, 
text = ''' 산에 산에 진달래꽃 피었습니다
진달래꽃 아름 따다 날 저뭅니다
한 잎 두 잎 꽃 뿌리며 돌아옵니다
뻐꾹새 먼 울음도 들려옵니다

 산에 산에 진달래꽃 피었습니다
진달래꽃 아름 따다 날 저뭅니다
산길은 봄 어스름 살살 내리고
저녁놀 서쪽 하늘 붉게 탑니다
''' 

""
"""
read() ->통째로 읽고 ->str
readline() -> 한줄만 읽고 -> str
readlines() ->  list
"""
```

```python
# 1번째 방법
f = open('newfile.txt', 'r')
a_line = f.read()
print(a_line)
f.close()
```

```python
# 2번째 방법
f = open('newfile.txt', 'r')
a_line = f.readlines()
for a in a_line:
    print(type(a_line))
    print(a, end='')
f.close()
```

```python
# 3번째 방법
f = open('newfile.txt', 'r')
a_line = f.readline()
while a_line:
    print(type(a_line))
    print(a_line,end='')
    a_line = f.readline()
f.close()
```

```python
# 4번째 방법
f = open('newfile.txt', 'r')
for i in f:
    print(type(i))
    print(i,end='')
f.close()
```

```python
# ban.txt 읽어서 딕셔너리로 저장
f = open('ban.txt', 'r')
ans = {}
for line in f:
    ans[line.split()[0]] = line.split()[1]
print(ans)
f.close()

# or 

f = open('ban.txt', 'r')
ans = {}
for line in f:
    (key, val) = line.split()
    ans[key] = val
print(ans)
f.close()
```

```python
# ban_student.txt 읽어서 다음과 같이 출력
f = open('ban_student.txt', 'r')
ans = {}
for line in f:
    (key, *val) = line.split()
    #(key, val) = line.split()[0], line.split()[1:]
    ans[key] = val
print(ans)
f.close()
```

```python
# stockcode.txt파일을 readline() 함수로 읽어서 다음과 같이 출력
f = open('stockcode.txt', 'r')
a_line = f.readline()
while a_line:
    print(a_line,end='')
    a_line = f.readline()
f.close()
```

```python
# stockcode.txt파일을 read() 함수로 읽어서 다음과 같이 출력
f = open('stockcode.txt', 'r')
a_line = f.read()
print(a_line)

f.close()
```

```python
# stockcode.txt파일을 readlines() 함수로 읽어서 다음과 같이 출력
# 2번째 방법
f = open('stockcode.txt', 'r')
a_line = f.readlines()
for a in a_line:
    print(a, end='')
f.close()
```

```python
# stockcode.txt.에서 '삼성 전자'의 거래소 번호를 출력하세요
f = open('stockcode.txt', 'r')
for line in f:
    if line.split()[1] == '삼성전자':
        print(line.split()[0])
```

```python
#입력받은 str를 text파일로 저장
def writtings():
    text = input('텍스트를 입력하세요: ')
    f = open('mydata.txt','w')
    f.write(text)
    f.close()
writtings()
```

```python
# 계속 입력받다가 enter만나면 끝나는.
def writtings_1():
    while True:
        text = input('텍스트를 입력하세요: ')
        if text == '':
            break
        else:
            f = open('mydata.txt','a')
            f.write(text)
            f.close()
writtings_1()
```

```python
# stockcode.txt를 읽어와서 data 복사해서 stockcode1.txt에 저장
f = open('stockcode.txt', 'r')
data = f.read()
g = open('stockcode1.txt', 'w')
g.write(data)
f.close()
g.close()
```

```python
# download.PNG파일 읽어와서 download1.PNG로 저장 /다 똑같은데 b : binary를 붙여줘서 ...
f = open('download_2.jpg', 'rb')
data = f.read()
g = open('download1.jpg', 'wb')
g.write(data)
f.close()
g.close()
```

```python
# 사이즈가 클 시에는... 100 바이트씩 나눠서 저장
buf_size = 100
f = open('download_2.jpg', 'rb')
g = open('download1.jpg', 'wb')
data = f.read(buf_size)
while data:
    g.write(data)
    data = f.read(buf_size)
f.close()
g.close()
```
